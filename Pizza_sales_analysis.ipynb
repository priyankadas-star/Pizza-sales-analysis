```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns

```

```python
df=pd.read_csv("C:/Users/priyanka das/OneDrive/Desktop/pizza_sales.csv")
print(df.head())
```

```python
print(df.info())
```

```python
print(df.shape)
```

```python
#total revenue
total_revenue=df['total_price'].sum()
print('Total Revenue:',total_revenue)
```

```python
#total orders
total_orders=df['order_id'].nunique()
print('Total Orders:',total_orders)
```

```python
#average order value
avg_order_value=total_revenue/total_orders
print('Average Order Value:',avg_order_value)
```

```python
#revenue by category
category_sales=df.groupby('pizza_category')['total_price'].sum()
print(category_sales)
```

```python
#revenue by size
size_sales=df.groupby('pizza_size')['total_price'].sum()
print(size_sales)
```

```python
#top10 pizzas
top_pizzas=df.groupby('pizza_name')['total_price'].sum().sort_values(ascending=False).head(10)
print(top_pizzas)
```

```python
#best selling pizza
best_selling = df.groupby("pizza_name")["quantity"].sum().sort_values(ascending=False).head(10)
print(best_selling)
```

```python
import matplotlib.pyplot as plt

category_sales.plot(kind="bar")

plt.title("Revenue by Category")
plt.ylabel("Revenue")
plt.show()
```
